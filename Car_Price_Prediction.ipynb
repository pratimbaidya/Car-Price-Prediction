### Cleaning the data

```python
df['year'].str.isnumeric()
```

```python
df = df[df['year'].str.isnumeric()]
```

```python
df['year'] = df['year'].astype(int)
```

```python
df.shape
```

```python
df.info()
```

```python
df['Price'] != 'Ask For Price'
```

```python
df = df[df['Price'] != 'Ask For Price']
```

```python
df['Price'] = df['Price'].str.replace(',', '').astype(int)
```

```python
df.info()
```

```python
df['kms_driven']
```

```python
df['kms_driven'] = df['kms_driven'].str.split(' ').str.get(0).str.replace(',', '')
```

```python
df = df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven'] = df['kms_driven'].astype(int)
```

```python
df.info()
```

```python
df = df[~df['fuel_type'].isna()]
```

```python
df.info()
```

```python
df['name'] = df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df
```

```python
df = df.reset_index(drop=True)
```

```python
df.describe()
```

```python
df[df['Price'] > 6e6]
```

```python
df = df[df['Price'] < 6e6].reset_index(drop=True)
```

```python
df.info()
```

## Model Building

```python
X = df.drop(columns='Price', axis = 1)
y = df['Price']
```

Imports

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, KFold, GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
ohe = OneHotEncoder(sparse_output= False)
ohe.fit(X[['name', 'company', 'fuel_type']])
```

```python
ohe.categories_
```

```python
lr = LinearRegression()
```

```python
X_train.head()
```

```python
col_tr = ColumnTransformer([('one_hot', OneHotEncoder(sparse_output= False, categories= ohe.categories_), ['name', 'company', 'fuel_type'])],
                           remainder='passthrough',
                           verbose_feature_names_out=False).set_output(transform = 'pandas')
```

```python
pipe = Pipeline([('col_tr', col_tr), ('lin_reg', lr)]).set_output(transform = 'pandas')
```

```python
pipe.fit(X_train, y_train)
```

```python
y_preds = pipe.predict(X_test)
```

```python
r2_score(y_test, y_preds)
```

#### Bootstrap Cross validation

```python
scores = []
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)
    lr = LinearRegression()
    pipe = Pipeline([('col_tr', col_tr), ('lin_reg', lr)]).set_output(transform = 'pandas')
    pipe.fit(X_train, y_train)
    y_preds = pipe.predict(X_test)
    scores.append(r2_score(y_test, y_preds))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= np.argmax(scores))
lr = LinearRegression()
pipe = Pipeline([('col_tr', col_tr), ('lin_reg', lr)]).set_output(transform = 'pandas')
pipe.fit(X_train, y_train)
y_preds = pipe.predict(X_test)
r2_score(y_test, y_preds)
```
